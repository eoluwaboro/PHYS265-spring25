```python
import numpy as np
import matplotlib.pyplot as plt
import pyspeckit

#load csv data
data = np.genfromtxt('silicon_spec_data.csv', delimiter=',', skip_header=1)
data = data[~np.isnan(data).any(axis=1)]
wavelength_nm = data[:, 0]
absorption_coeff = data[:, 1] #in cm^-1
energy = 1240 / wavelength_nm #in eV

#create spectrum for PySpecKit
sp = pyspeckit.Spectrum(xarr=energy, data=absorption_coeff, unit='1/cm', xunit='eV')
sp.plotter()

sp.specfit(fittype='gaussian', guesses=[absorption_coeff.max(), 1.1, 0.05])
sp.specfit.plot_fit()

sp.plotter.axis.set_xlabel('Photon Energy(eV)')
sp.plotter.axis.set_ylabel('Absorption Coefficient (cm$^{-1}$)')
sp.plotter.figure.suptitle('Silicon Absorption Spectrum with Gaussian Fit')
plt.tight_layout()
```
