```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import LogNorm

G = 6.67e-11
M_earth = 5.9e24
M_moon = 7.3e22
dE2M = 3.8e8

x_earth, y_earth = 0, 0
x_moon, y_moon = dE2M / np.sqrt(2), dE2M / np.sqrt(2)

def F(G, M, m, xM, yM, X, Y):
    rx = X - xM
    ry = Y - yM
    r = np.sqrt(rx**2 + ry**2)
    r[r == 0] = 1e-10  # To avoid division by zero at the mass point
    F_mag = -G * (M * m) / r**2
    Fx = F_mag * (rx / r)
    Fy = F_mag * (ry / r)
    return Fx, Fy

x = np.linspace(-1.5*dE2M, 1.5*dE2M, 50)
y = np.linspace(-1.5*dE2M, 1.5*dE2M, 50)
X, Y = np.meshgrid(x, y)

Fx_earth, Fy_earth = F(G, M_earth, M_moon, x_earth, y_earth, X, Y)
Fx_moon, Fy_moon = F(G, M_moon, M_earth, x_moon, y_moon, X, Y)

Fx = Fx_earth + Fx_moon
Fy = Fy_earth + Fy_moon
z = np.sqrt(Fx**2 + Fy**2)

fig, ax = plt.subplots()
ax.streamplot(X, Y, Fx, Fy, color='k', linewidth=1, density =1.25)
c = ax.pcolormesh(X, Y, z, cmap='viridis',norm = LogNorm(vmin=1e-10, vmax=z.max()))
fig.colorbar(c, ax=ax,label = 'Force Magnitude (N)')
ax.plot(x_earth, y_earth, 'bo', label='Earth', markersize=10)
ax.plot(x_moon, y_moon, 'o',color ='grey', label='Moon', markersize=5)

ax.set_xlabel('x (m)')
ax.set_ylabel('y (m)')
ax.set_aspect('equal')
ax.set_title('Gravitational Force Streamplot')
fig.tight_layout()
```
