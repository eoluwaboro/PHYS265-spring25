```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import LogNorm


# gravitational constant = 6.67E-11 m^3/kg/s^2
# gravitational acceleration = 9.81 m/s^2
# Mass of earth = 5.9E24 kg
# mass of moon = 7.3E22 kg
# mass of apollo module = 5500 kg
# Radius of earth 6378 km
# radius of the moon =  1737 km
# distance from earth to moon = 3.8E8 m
# exhaust velocity = 2.4E3
# burn rate = 1.3E4
#wet mass = 2.4E3
# Dry mass = 7.5E5

G = 6.67e-11

M_earth = 5.9e24
M_moon = 7.3e22
dE2M = 3.8e8

x_earth, y_earth = 0,0
x_moon, y_moon = dE2M / np.sqrt(2), dE2M / np.sqrt(2)

def phi(G, M, xM, yM,x, y):
    r = np.sqrt((x-xM)**2 + (y-yM)**2)
    return -G*M/r

x = np.linspace(-1.5*dE2M, 1.5*dE2M, 50)
y = np.linspace(-1.5*dE2M, 1.5*dE2M, 50)
X,Y = np.meshgrid(x,y)

Gpot_earth = phi(G,M_earth,x_earth,y_earth,X,Y)
Gpot_moon = phi(G,M_moon,x_moon,y_moon,X,Y)
Gpot_abs = np.abs(Gpot_earth + Gpot_moon)    

fig1, ax1 = plt.subplots()
c = ax1.pcolormesh(X,Y, Gpot_abs, cmap = 'viridis',norm = LogNorm(vmin=1e4, vmax=1e9))
ax1.set_xlabel('x(m)')
ax1.set_ylabel('y(m)')
ax1.set_aspect('equal')
ax1.set_title('2D Gravitational Potential Field')
fig1.colorbar(c,label='Gravitational Potential (J/kg)')
ax1.contour(X, Y, Gpot_abs, levels=20, colors='white', linewidths=0.5)

#ax1.plot(x_earth, y_earth, 'bo', label = 'Earth', markersize = 10)
#ax1.plot(x_moon, y_moon, 'ro', label = 'Moon', markersize = 5)




```
