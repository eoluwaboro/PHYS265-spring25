```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad


# gravitational constant = 6.67E-11 m^3/kg/s^2
# gravitational acceleration = 9.81 m/s^2
# Mass of earth = 5.9E24 kg
# mass of moon = 7.3E22 kg
# mass of apollo module = 5500 kg
# Radius of earth 6378 km
# radius of the moon =  1737 km
# distance from earth to moon = 3.8E8 m
# exhaust velocity = 2.4E3
# burn rate = 1.3E4
#wet mass = 2.4E3
# Dry mass = 7.5E5

G = 6.67e-11
def phi(G, M, xM, yM,x, y):
    r = np.sqrt((x-xM)**2 + (y-yM)**2)
    return -G*M/r
    
M = 5.9e24
xM, yM = 0,0
dE2M = 3.8e8
r = np.linspace(6e6,dE2M,1000)


x = np.linspace(-1.5*dE2M, 1.5*dE2M, 500)
y = np.linspace(-1.5*dE2M, 1.5*dE2M, 500)
X,Y = np.meshgrid(x,y)

Gpot = phi(G,M,xM,yM,r,0)
Gpot_abs = np.abs(Gpot)  

fig1,ax1 = plt.subplots()
ax1.plot(r*1.5,Gpot_abs, label='|Gravitational Potential|')
ax1.set_xlabel('Distance from Earth to Moon(m)')
ax1.set_ylabel('Potential (J/kg)')
ax1.set_yscale('log')
ax1.set_title('Gravitational Potential vs. Distance')
ax1.grid(True)
ax1.legend()

Gpot = phi(G,M,xM,yM,X,Y)
Gpot_abs = np.abs(Gpot)  

fig2, ax2 = plt.subplots()
c = ax2.pcolormesh(X,Y, Gpot_abs, cmap = 'viridis')
ax2.set_xlabel('x(m)')
ax2.set_ylabel('y(m)')
ax2.set_aspect('equal')
ax2.set_title('2D Gravitational Potential Field')
fig2.colorbar(c,label='Gravitational Potential (J/kg)')
fig.tight_layout();



```
